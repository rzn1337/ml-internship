```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Creating the dataset
np.random.seed(0)
num_samples = 100

square_footage = np.random.randint(500, 3500, num_samples)
bedrooms = np.random.randint(1, 6, num_samples)
bathrooms = np.random.randint(1, 4, num_samples)

# Generating prices with some random noise
prices = (square_footage * 200) + (bedrooms * 10000) + (bathrooms * 5000) + np.random.randint(-20000, 20000, num_samples)

data = pd.DataFrame({
    'Square Footage': square_footage,
    'Bedrooms': bedrooms,
    'Bathrooms': bathrooms,
    'Price': prices
})
```

```python
X = data[['Square Footage', 'Bedrooms', 'Bathrooms']]
y = data['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Plot the results
plt.scatter(y_test, y_pred)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual vs Predicted Prices')
plt.show()
```
